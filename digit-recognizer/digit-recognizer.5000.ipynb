```python
train_read = pd.read_csv('../input/digit-recognizer/train.csv')
image=train_read.iloc[0:5000,1:]
label=train_read.iloc[0:5000,:1]
train_images,test_image,train_label,test_label=train_test_split(image,label,train_size=0.8,random_state=0)

i=1
img_draw=train_images.iloc[i].to_numpy()
img_draw=img_draw.reshape((28,28))
plt.imshow(img_draw,cmap='gray')
plt.title(train_label.iloc[i,0])

#plot frequency:numbers in lables
plt.hist(train_images.iloc[i])


# training our module and find what is its efficency

clf=svm.SVC()
clf.fit(train_images,train_label.values.ravel())
clf.score(test_image,test_label)

test_read=pd.read_csv('../input/digit-recognizer/test.csv')

results=clf.predict(test_read[0:5000])

df = pd.DataFrame(results)
df.index.name='ImageId'
df.index+=1
df.columns=['Label']
df.to_csv('results.csv', header=True)
```
